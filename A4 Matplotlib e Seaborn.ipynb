# Basic Errorbars

Le barre di errore (errorbars) sono rappresentazioni grafiche della variabilità dei dati e vengono utilizzate sui grafici per indicare l'errore o l'incertezza in una misurazione riportata

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

# visualizzare una funzione tridimensionale

Supponiamo di voler visualizzare la funzione :

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50) # creiamo l'asse x dello spazio tridimensionale
y = np.linspace(0, 5, 40) # creiamo l'asse y dello spazio tridimensionale

X, Y = np.meshgrid(x, y)  # creazione di una griglia bidimensionale
Z = f(X, Y)               # creazione dell'asse Z con la funzione da visualizzare

plt.contour(X, Y, Z, colors='black');
```

```python
le linee tratteggiate sono valori negativi mentre quelle piene sono positivi
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy'); # mappa a due colori 
```

```python
# possiamo anche far riempire i contorni con dei colori e stampare la color bar
plt.contourf(X, Y, Z, 20, cmap='RdGy') 
plt.colorbar();
```

```python
# la funzione imshow disegna il tutto come immagine
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
```

```python
# imshow permette di avere anche un canale alpha
# con le opportune accortezze è possibile unire due grafici

contours = plt.contour(X, Y, Z, 3, colors='black')  # disegno i contorni
plt.clabel(contours, inline=True, fontsize=8)  # con le etichette

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',cmap='RdGy', alpha=0.5) # e sovrappongo l'immagine con alpha 0.5
plt.colorbar();  # gli dò anche la colorbar
```

# Kernel density estimation - KDE

La Kernel density estimation serve per rendere le transizioni più morbide

```python
from scipy.stats import gaussian_kde

# creo i dati
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[-3.5, 3.5, -6, 6],
           cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```

# Subplots

plt.axes accetta un argomento opzionale che è un elenco di quattro numeri nel sistema di coordinate della figura. 
Questi numeri rappresentano [sinistra, fondo, larghezza, altezza] nel sistema di coordinate della figura, che va da 0 in basso a sinistra della figura a 1 in alto a destra della figura

plt.subplot() usa plt.axes e fà un pò di lavoro per noi. 
plt.subplot() crea una singola sottotrama all'interno di una griglia. 
Questo comando accetta tre argomenti interi: il numero di righe, il numero di colonne e l'indice del grafico da creare in questo schema, che va da in alto a sinistra a in basso a destra


```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
           fontsize=18, ha='center')
    
# Il comando plt.subplots_adjust è usato per regolare la spaziatura tra questi grafici    
```

Il metodo plt.GridSpec() usa plt.subplot(). 
Ovvero è semplicemente una comoda interfaccia che viene riconosciuta dal comando plt.subplot(). 
Ad esempio, una gridspec per una griglia di due righe e tre colonne con uno spazio di larghezza e altezza specificati ha il seguente aspetto:

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

Questo tipo di allineamento flessibile della griglia ha un'ampia gamma di usi. 
E' usato spesso quando vengono creati grafici a istogramma multiasse come quelli mostrati qui:

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d') # projection='3d' abilita il plot in 3D
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

### Plot contorni 3D
Analogamente ai grafici di contorno che abbiamo esplorato prima, mplot3d contiene strumenti per creare grafici di rilievo tridimensionali utilizzando gli stessi input.

```python
# funzione da visualizzare
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

# creo gli indici per x e y
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)  # creo la griglia
Z = f(X, Y) # calcolo Z con la funzione definita sopra
fig = plt.figure() # creo la figura
ax = plt.axes(projection='3d') # preparo l'area di disegno in 3d
ax.contour3D(X, Y, Z, 50, cmap='binary') # plotto
ax.set_xlabel('x') # applico le label
ax.set_ylabel('y')
ax.set_zlabel('z')

# Altri tipi di plotting
#ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none') # superfici colorate
#ax.plot_wireframe(X, Y, Z, color='black') # plotto wireframe # wireframe

# Tricks
#ax.view_init(60, 35) # aggiusto elevazione e grado del punto di vista  
```

E se invece di un'estrazione pari da una griglia cartesiana o polare, avessimo invece una serie di estrazioni casuali?
Potremmo creare un grafico a dispersione dei punti per avere un'idea della superficie da cui stiamo campionando:

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5);
```

per creare una superficie possiamo usare ax.plot_trisurf, che trova un insieme di triangoli formati tra punti adiacenti

```python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z,
                cmap='viridis', edgecolor='none');
```
