# Dati e Pattern

Nel machine learning il comportamento degli algoritmi non è pre-programmato ma appreso dai **dati** stessi. <br>
Quindi i dati sono un ingrediente fondamentale che determinano il comportamento dell'algoritmo. <br>
Con queste condizioni si capisce l'importanza di avere dei dati puliti e veritieri. Altrimenti sarebbe come programmare un codice di tipo vecchio con un'analisi errata o non completa. 

Un **pattern** possiamo definirlo come un entità vagamente definita a cui può essere dato un nome. <br>
In italiano può essere tradotto come : forma, campione, esempio, modello, ecc. <br>
Ad esempio un pattern può essere un volto, un carattere scritto a mano, un’impronta digitale, un segnale sonoro, un
frammento di testo, l’andamento di un titolo di borsa.

Il **Pattern Recognition** è la disciplina che studia il riconoscimento dei pattern (non solo con tecniche di learning ma anche con algoritmi pre-programmati). Ovviamente i principali algoritmi per il pattern recognition fanno parte del dominio del Machine Learning



## Tipi di Pattern

Dato che esistono vari tipi di pattern, esistono vari tipi di algoritmi che sono specializzati nel riconoscimento di specifici pattern.

I principali tipi di pattern sono Numerici, Categorici e Sequenze.
 
**Numerici** : valori associati a caratteristiche misurabili o conteggi. <br> 
Tipicamente continui (ma anche discreti, es. interi), in ogni caso soggetti a ordinamento.
Rappresentabili naturalmente come vettori numerici nello spazio multidimensionale.
L’estrazione di caratteristiche da segnali (es., immagini, suoni) produce vettori numerici detti anche feature vectors.
Es. Persona: [altezza, circonferenza toracica, circonferenza fianchi, lunghezza del piede]
Principale tipologia di dati considerata in questo corso.

**Categorici** : valori associati a caratteristiche qualitative e alla presenza/assenza di una caratteristica (yes/no value).
Non «semanticamente» mappabili in valori numerici.
Es. Persona: [sesso, maggiorenne, colore occhi, gruppo sanguigno].
Talvolta soggetti a ordinamento (ordinali): es. temperatura ambiente: alta, media o bassa.
Normalmente gestiti da sistemi a regole e alberi di classificazione.
Molto utilizzati nell’ambito del data mining, spesso insieme a dati numerici (mixed).

Sequenze: pattern sequenziali con relazioni spaziali o temporali.
Es. uno stream audio (sequenza di suoni) corrispondente alla pronuncia di una parola, una frase (sequenza di parole) in
linguaggio naturale, un video (sequenza di frame).
Spesso a lunghezza variabile 
La posizione nella sequenza e le relazioni con predecessori e successori sono importanti.
Critico trattare sequenze come pattern numerici.
Allineamento spaziale/temporale, e «memoria» per tener conto del passato.

Approcci:
Dynamic Time Warping (DTW)
Hidden Markov Models (HMM)
Recurrent Neural Networks (RNN)
Long Short-Term Memory (LSTM)


Altri dati strutturati: 
output organizzati in strutture complesse quali alberi e grafi.
Applicazioni in Bioinformatics, Natural Language Processing, Speech recognition, ecc.
Esempio nella traduzione di una frase in linguaggio naturale, l’output desiderato è l’insieme dei parse tree plausibili.
Approcci: Structured SVMs, Bayesian Networks, HMM 

**Classificazione**: **assegna una classe a un pattern**.
Necessario apprendere una funzione capace di eseguire il mapping dallo spazio dei pattern allo spazio delle classi
Si usa spesso anche il termine riconoscimento. 
Nel caso di 2 sole classi si usa il termine binary classification, con più di due classi multi-class classification.
Classe: insieme di pattern aventi proprietà comuni.
Es. i diversi modi in cui può essere scritto a mano libera il carattere A.
Il concetto di classe è semantico e dipende strettamente dall’applicazione:
- 21 classi per il riconoscimento di lettere dell’alfabeto
- 2 classi per distinguere le lettere dell’alfabeto italiano da quello cirillico
Esempi di problemi di classificazione:
Spam detection
Credit Card fraud detection
Face recognition
Pedestrian classification
Medical diagnosys
Stock Trading

**Regressione**: **assegna un valore continuo a un pattern**.
Utile per la predizione di valori continui.
Risolvere un problema di regressione corrisponde ad apprendere una funzione approssimante delle coppie «input,output» date.
Quale tipo di funzione? Con quale grado di regolarità?
Esempi di problemi di regressione:

**Modello**: un modello di machine learning può essere una rappresentazione matematica di un processo del mondo reale. 
Per generare un modello di machine learning si devono fornire dati di training a un algoritmo di machine learning. Tale algoritmo addestrerà il modello facendogli apprendere la relazione insita nei dati.

**Features**: le features (caratteristiche in italiano), a volte chiamati anche **attributi**, sono variabili che fungono da input nel sistema. <br>
Il numero di features è chiamato dimensione delle features. <br>
Ad esempio un regressione lineare che predice il valore di una casa, potrebbe avere come features la grandezza della casa, la posizione, la presenza del giardino, etc... <br>
Attenzione: la feature è la variabile non il suo valore. Facendo una analogia con un foglio excel, la feature è la colonna non i valori o il valore della colonna.

**Label**: la label è l'output finale del modello.<br> 
Ad esempio un regressione lineare che predice il valore di una casa ha come label il valore in migliaia di euro della casa.
Attenzione: come per la feature, anche la label è la variabile non il suo valore.

**Datapoint**: i datapoint sono i dati reali delle features per un singolo evento<br>
Attenzione: facendo la stessa analogia del foglio excel corrisponderebbe ad una riga delle colonne delle features.

**Target**: il target è il valore y rispetto alle variabili di input x. <br>
In altre parole, dato un training set, i target sono i valori y che sono usati per allenare il modello

La differenza tra la label e il target è che le label sono i risultati del modello mentre il target è la variabile y su cui dobbiamo fare training

**Overfitting**: è un parametro importantissimo dell'apprendimento automatico. <br>
Indica quanto, il modello addestrato, riesce a predire correttamente l'ouput a partire dai nuovi dati. <br>
La generalizzazione funziona meglio se il segnale o il campione utilizzato come dati di addestramento ha un rapporto segnale/rumore elevato. <br>
In altre parole un modello è in overfitting se si adatta troppo bene ai dati di training e c'è una scarsa generalizzazione dei nuovi dati.

**Parametri e IperParametri** : i parametri sono quelli che sono appresi dalla macchina come pesi e bias. Ad esempio nel caso di una regressione lineare (y = ax+b ) sono a e b. <br>
Gli iperparametri sono quelli che forniamo al framework per la definizione del modello stesso. Ad esempio: numero di nodi e livelli nascosti, funzioni di input, tasso di apprendimento (alpha), funzione di attivazione ecc. 

In parole ancora più semplificate :

Nel machine learning ci sono tre tipi di reti neurali.
 - apprendimento supervisionato : impara da input e output
 - apprendimento non supervisionato : impara solo dagli input
 - apprendimento con rinforzo : impara con una funzione di guadagno
 
 Ora vediamo i tre tipi rispetto ad un dataset di esempio contenente bici e auto

```python
Image(filename = "images\DatasetExampleML.PNG", width=656, height=918)
```

### Apprendimento supervisionato
Nell'apprendimento supervisionato noi diamo al modello, durante la fase di training, i dati di input e i dati di output.<br>
Quindi il modello apprende usando dati "etichettati"<br>
L'algoritmo di IA farà imparare al modello la correlazione tra l'input e l'output. <br>
Quindi per ogni osservazione sono noti valori di input e valore di output.<br>
Esistono due tipi di apprendimento supervisionato :
 - Regressione : output continuo
 - Classificazione : output discreto 

```python
Image(filename = "images\ClassRegr.png", width=656, height=918)
```

Nell'apprendimento supervisionato mostriamo prima tutte le biciclette (datapoint/input) dicendo che sono biciclette (target) ma senza indicare le loro caratteristiche. Poi facciamo lo stesso per le auto. <br>
Una volta addestrato il modello gli sottoponiamo una bici (input) e gli chiediamo di dirci cos'è (label).

### Apprendimento non supervisionato

Nell'apprendimento non supervisionato il modello apprende usando dati non etichettati. <br>
Il modello deve apprendere usando solo gli input (datapoint) e non l'output (target). <br>
Normalmente si usa apprendimento non supervisionato per tipologie di problemi più complessi di quelli risolvibili con l'apprendimento supervisionato.

Dando il dataset di biciclette e auto al modello senza dirgli nulla dell'output aspettato, ovvero senza dirgli quali sono le bici e quali le auto, il modello classificherà in autonomia gli oggetti dati.

L'ovvio **pro** dell'apprendimento non supervisionato è il fatto di non dare l'output (target) del rispettivo input (datapoint). Molte volte si usa l'apprendimento non supervisionato quando non abbiamo i dati target.

I **contro** dell'apprendimento non supervisionato è il fatto che il modello potrebbe : 
 - trovare più o meno categorie che ci aspetteremmo (ad es. 3 o 4 categorie basate anche su altre features)
 - trovare categorie diverse da quelle che ci aspetteremmo (ad es. categorie per colore)

### Apprendimento con rinforzo

Nell'apprendimento con rinforzo l'apprendimento è guidato da una "funzione di guadagno" (o di costo) che viene aumentata o diminuita ogni volta che il modello raggiunge uno stato positivo o negativo. <br>
Quindi è come avere uno stato di bonus/malus dipendente dallo stato del modello.

## Derivate e rappresentazione geometrica

Dal punto di vista geometrico, una derivata misura la pendenza del grafico di una funzione in un punto x0 ossia il **coefficiente angolare** della retta tangente nel punto x0.

In geometria analitica il coefficiente angolare (in lingua inglese slope, pendenza) di una retta non verticale nel piano cartesiano è il coefficiente **m** che compare nell'equazione: <br>
 - y=**m**x+q

Per creare valore all'azienda, dobbiamo seguire quattro fasi.
 - La prima fase è quella **descrittiva**, (descriptive) correggiamo e puliamo i dati per poi iniziare a investigare
 - La seconda fase è quella **diagnostica**, (diagnostic) investighiamo nei dati per trovare le cause e i pattern utili
 - La seconda fase è quella **predittiva**, (predittiva) dove usiamo il modello per capire il futuro
 - La terza fase è quella **prescrittiva**, (prescriptive) dove cerchiamo di capire, modello alla mano, cosa dovrebbe fare l'azienda e quali decisioni dovrebbe prendere


**Fase descrittiva** : (cosa è successo?)
 - Abbiamo dei dati grezzi
 - Pulizia dati
 - Report e grafici

**Fase diagnostica** :  (perchè è avvenuto?)
 - Cerco le cause
 - Cerco esempi e ricorrenze

**Fase predittiva** :  (cosa è probabile che accada?)
 - Creo il modello con il giusto algoritmo di ML
 - Creo la giusta strategia aziendale

**Fase prescrittiva** :  (cosa devo fare?)
 - Decidere cosa devo fare una volta che riesco a 'prevedere' il futuro
